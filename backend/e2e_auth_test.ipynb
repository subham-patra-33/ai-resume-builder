```python
import urllib.request
import urllib.error
import json
import time

def post(url, data):
    req = urllib.request.Request(url, data=json.dumps(data).encode('utf-8'), headers={'Content-Type': 'application/json'}, method='POST')
    try:
        with urllib.request.urlopen(req, timeout=10) as resp:
            return resp.status, json.loads(resp.read().decode('utf-8'))
    except urllib.error.HTTPError as err:
        return err.code, json.loads(err.read().decode('utf-8'))
    except Exception as err:
        return None, {'error': str(err)}

email = f'e2e-user-{int(time.time())}@example.com'
print('email', email)
status, body = post('http://localhost:4000/api/auth/send-otp', {'email': email})
print('sendOtp', status, body)
print('If the OTP was delivered with SMTP fallback, check backend terminal output for the code.')
```
